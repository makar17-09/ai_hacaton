```python
from training_score_predict import PredictData
from tensorflow import keras
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import mean_squared_error
```

```python
train="Task/train_all.csv"
test="Task/test_all.csv"
metod = "NCA"
```

```python
sample = PredictData(train, test)
new_test_data = sample.data_conversion(metod, 1)
```

```python
X_train = sample.X_train
y_train = sample.y_train
model = keras.Sequential(
    [
    keras.layers.Dense(
        X_train.shape[-1], activation="relu", input_shape=(X_train.shape[-1],)
    ),
    keras.layers.Dense(1500, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-4)),
#     keras.layers.Dropout(0.1),
    keras.layers.Dense(999, activation="relu"),
    keras.layers.Dense(500, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-4)),
    keras.layers.Dense(1, activation="sigmoid"),
    ]
    )


##########################################

metrics = [
    keras.metrics.FalseNegatives(name="fn"),
    keras.metrics.FalsePositives(name="fp"),
    keras.metrics.TrueNegatives(name="tn"),
    keras.metrics.TruePositives(name="tp"),
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
    ]



model.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="categorical_crossentropy", metrics=metrics
    )
```

```python
y_train
```

```python
model.fit(
        X_train,
        y_train,
        batch_size = 1000,
        epochs = 500,
    )
```
