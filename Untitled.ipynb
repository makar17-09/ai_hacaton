```python
import pandas as pd
from tqdm import tqdm
from urllib.request import urlopen
from urllib.parse import quote
from rdkit import Chem
import numpy as np

from gensim.models import word2vec
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("train_new.csv")
```

```python
from rdkit.Chem.Descriptors import ExactMolWt

mol_wt = []

for mol in tqdm(data["Smiles"]):
    m = Chem.MolFromSmiles(mol)
    mol_wt.append(ExactMolWt(m)) #weight of the molecule
```

```python
data["Weight"] = mol_wt
```

```python
r = []
for a in data["Active"]:
    r.append(1 if a else 0)
y = pd.Series(r, copy=False)
```

```python
data = data[['Smiles', "Weight"]]
```

```python
data
```

```python
y
```

```python
all_smiles = []
for s in data["Smiles"]:
    all_smiles.append([s])
```

```python
print("Training model...")

%time model = word2vec.Word2Vec(all_smiles, min_count=1)
```

```python
X_emb = []
for word in model.wv.key_to_index:
    X_emb.append(model.wv[word])
```

```python
X = pd.DataFrame(X_emb, copy=False)
```

```python
X["Weight"] = mol_wt
```

```python
X["Weight"] = X["Weight"] / 5557
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

```python
X_train
```

```python
from tensorflow import keras

hid_size = 101
model = keras.Sequential(
    [
        keras.layers.Dense(
            hid_size, activation="relu", input_shape=(X_train.shape[-1],)
        ),
        keras.layers.Dense(50, activation="relu"),
        keras.layers.Dense(50, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid"),
    ]
)
```

```python
metrics = [
    keras.metrics.FalseNegatives(name="fn"),
    keras.metrics.FalsePositives(name="fp"),
    keras.metrics.TrueNegatives(name="tn"),
    keras.metrics.TruePositives(name="tp"),
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
]
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="binary_crossentropy", metrics=metrics
)
```

```python
model.fit(
    X_train,
    y_train,
    batch_size=2048,
    epochs=1000,
    verbose=2,
)
```

```python
y_pred = model.predict(X_test)
```

```python
y_predict = []
for y in y_pred:
    int_y = 1 if y > 0.4 else 0
    y_predict.append(int_y)
```

```python
accuracy_score(y_test, y_predict)
```

```python
confusion_matrix(y_test, y_predict)
```
