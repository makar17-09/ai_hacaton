```python
import pandas as pd
import numpy as np

from sklearn.neighbors import NeighborhoodComponentsAnalysis
from catboost import CatBoostClassifier
from tensorflow import keras
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("Task/train_all.csv")
data = data.loc[:, "Smiles":]
data_test = pd.read_csv("Task/test_all.csv")
```

```python
name_col = data.columns.values # получим наименования столбцов, из них удалим первый и второй, т.к. первый в обучение
data_col = np.delete(name_col, [0, 1]) #не подходит, а второй это правильные метки
X = data[data_col]

r = []
for a in data["Active"]:
    r.append(1 if a else 0)
y = pd.Series(r, copy=False) #правильные метки должны быть 0 и 1, а не True и False

conv = StandardScaler()
conv.fit(X, y)
data_train_processed = conv.transform(X)
data_test_processed = conv.transform(data_test[data_col])

X_train, X_test, y_train, y_test = train_test_split(data_train_processed, y, test_size=0.3, stratify=y)
```

```python
clf = CatBoostClassifier(iterations=1, learning_rate=0.00005, depth=8)
clf.fit(X_train, y_train)
y_predc = clf.predict(X_test)
a = accuracy_score(y_test, y_predc)
cm = confusion_matrix(y_test, y_predc)
print(f"Accuracy {a} \nConfusion matrix\n {cm}")
```

```python
model = keras.Sequential(
[
    keras.layers.Dense(
        X_train.shape[-1], activation="relu", input_shape=(X_train.shape[-1],)
    ),
    keras.layers.Embedding(input_dim=1000, output_dim=64),
    keras.layers.GRU(256, return_sequences=True),
#     keras.layers.SimpleRNN(128),
    keras.layers.Dense(1, activation="sigmoid"),
]
)
```

```python
metrics = [
    keras.metrics.FalseNegatives(name="fn"),
    keras.metrics.FalsePositives(name="fp"),
    keras.metrics.TrueNegatives(name="tn"),
    keras.metrics.TruePositives(name="tp"),
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
]
```

```python
model.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="binary_crossentropy", metrics=metrics
)
```

```python
model.fit(
    X_train,
    y_train,
    batch_size = 2048,
    epochs = 500,
)
```

```python
y_pred_1 = model.predict(X_test)
```

```python
model_2 = keras.Sequential(
[
    keras.layers.Dense(
        X_train.shape[-1], activation="relu", input_shape=(X_train.shape[-1],)
    ),
    keras.layers.Dense(300, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-4)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(666, activation="relu"),
    keras.layers.Dense(444, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-3)),
    keras.layers.Dense(1, activation="sigmoid"),
]
)
```

```python
model_2.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="binary_crossentropy", metrics=metrics
)

model_2.fit(
    X_train,
    y_train,
    batch_size = 2048,
    epochs = 500,
)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
model_3 = keras.Sequential(
[
    keras.layers.Dense(
        X_train.shape[-1], activation="relu", input_shape=(X_train.shape[-1],)
    ),
    keras.layers.Dense(400, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-4)),
    keras.layers.Dense(400, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-3, l2=1e-2)),
    keras.layers.Dense(400, activation="relu", kernel_regularizer=keras.regularizers.l1_l2(l1=1e-4, l2=1e-4)),
    keras.layers.Dense(1, activation="sigmoid"),
]
)
```

```python
model_3.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="binary_crossentropy", metrics=metrics
)

model_3.fit(
    X_train,
    y_train,
    batch_size = 2048,
    epochs = 500,
)
```
